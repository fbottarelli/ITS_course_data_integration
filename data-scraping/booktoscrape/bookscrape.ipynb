Ora prendiamo i dati più specifici di ogni libro guardando nella pagina del libro.

```python
# ciclo come sopra ma per ogni articolo andiamo a cercare il link e andiamo a fare il parsing della pagina del libro
import re

dataset = []
for i in range(1, 2):
    url = f"https://books.toscrape.com/catalogue/page-{i}.html"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    articles = soup.find_all("article")
    for article in articles:
        link = article.h3.a["href"]
        book_page = requests.get(f"https://books.toscrape.com/catalogue/{link}")
        book_soup = BeautifulSoup(book_page.text, "html.parser")
        row = {
            "Title": article.h3.a["title"],
            "Link": article.h3.a["href"],
            "Price": article.find("p", class_="price_color").text.replace("Â£", "").strip(),
            "Availability": article.find("p", class_="instock availability").text.strip()
        }
        if row["Availability"]=="In stock":
            number_of_book_available = book_soup.find("p", class_="instock availability").text.strip()
            # print(number_of_book_available)
            books_available = int(re.findall(r'\d+', number_of_book_available)[0])
            # aggiugiamo il numero di libri disponibili alla riga del df
            row["Number of books available"] = books_available
        # ora aggiungiamo le altre informazioni che ci interessano dalla pagina del libro
        description = book_soup.find("div", id="product_description")
        if description:
            description = description.find_next_sibling("p").text
        else:
            description = None
        row["Description"] = description
        product_information = book_soup.find("table", class_="table table-striped")
        for tr in product_information.find_all("tr"):
            key = tr.th.text
            value = tr.td.text.replace("Â£", "")
            row[key] = value
        genres = book_soup.find("ul", class_="breadcrumb").find_all("a")[2]
        row["Genre"] = genres.text

        dataset.append(row)
df = pd.DataFrame(dataset)
# salviamo locale il df
df.to_csv("books.csv", index=False)
df
    
```

```python
from dotenv import load_dotenv
import os
import requests
load_dotenv()
# openAI_key = os.getenv("OPENAI_KEY")
```

```python
from openai import OpenAI
client = OpenAI()

# Supponiamo che 'book_descriptions' sia una lista di descrizioni di libri in inglese ottenuta dal tuo web scraping
book_descriptions = ["A detailed exploration of the life of Leonardo Da Vinci.", "An insightful look into the complexities of artificial intelligence."]

# Ciclo per tradurre ciascuna descrizione in italiano
for description in book_descriptions:
    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "Translate English text into Italian and summarize it in a text max 20 words"},
            {"role": "user", "content": description}
        ]
    )

    # Stampa la traduzione in italiano
    translated_text = completion.choices[0].message.content
    print(translated_text)
```

```python
# Creazione di un'istanza del client OpenAI
client = OpenAI()

dataset = []
for i in range(1, 2):  # qui ciclo solo per la prima pagina per semplicità
    url = f"https://books.toscrape.com/catalogue/page-{i}.html"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    articles = soup.find_all("article")
    for article in articles:
        link = article.h3.a["href"]
        book_page = requests.get(f"https://books.toscrape.com/catalogue/{link}")
        book_soup = BeautifulSoup(book_page.text, "html.parser")
        row = {
            "Title": article.h3.a["title"],
            "Link": article.h3.a["href"],
            "Price": article.find("p", class_="price_color").text.replace("Â£", "").strip(),
            "Availability": article.find("p", class_="instock availability").text.strip()
        }

        if row["Availability"]=="In stock":
            number_of_book_available = book_soup.find("p", class_="instock availability").text.strip()
            books_available = int(re.findall(r'\d+', number_of_book_available)[0])
            row["Number of books available"] = books_available

        description = book_soup.find("div", id="product_description")
        if description:
            description_text = description.find_next_sibling("p").text
            # Traduzione della descrizione in italiano
            completion = client.chat.completions.create(
                model="gpt-3.5-turbo",
                messages=[
                    {"role": "system", "content": "Traduci il testo inglese in italiano e riassumi in poche parole (massimo 30 parole)"},
                    {"role": "user", "content": description_text}
                ]
            )
            translated_description = completion.choices[0].message.content
            row["Description"] = translated_description
        else:
            row["Description"] = None

        product_information = book_soup.find("table", class_="table table-striped")
        for tr in product_information.find_all("tr"):
            key = tr.th.text
            value = tr.td.text.replace("Â£", "")
            row[key] = value

        genres = book_soup.find("ul", class_="breadcrumb").find_all("a")[2]
        row["Genre"] = genres.text
        dataset.append(row)

df = pd.DataFrame(dataset)
# Salvataggio del dataframe in un file CSV
df.to_csv("books.csv", index=False)
```
